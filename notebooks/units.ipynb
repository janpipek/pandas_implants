```python
temps = pd.DataFrame([
    ["Prague", 23],
    ["Ostrava", 27],
    ["Nairobi", 31],
    ["Cairo", 47],
    ["Johannesburg", 13]
], columns=["city", "temp"])
temps
```

```python
temps["temp"] = temps["temp"].astype("unit[deg_C]")
temps
```

```python
temps.dtypes
```

```python
temps["temp_f"] = temps["temp"].units.to("deg_F")
temps["temp_K"] = temps["temp"].units.to("K")
temps
```
